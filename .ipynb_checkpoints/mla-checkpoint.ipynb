## Algoritmo

Primeiro, realizamos uma operação *drop* nos tópicos ou linguagens que só são usados uma vez em todo o dataset. Não será possível criar grupos com eles, então para regras de associação são irrelevantes

```python
# retorna todas as colunas com somatório 1
df_topics_cols_drop = df_topics.loc[:, df_topics.sum() == 1].columns
df_languages_cols_drop = df_languages.loc[:, df_languages.sum() == 1].columns

# dropa essas colunas
df_topics = df_topics.drop(columns=df_topics_cols_drop)
df_languages = df_languages.drop(columns=df_languages_cols_drop)
```

```python
df_topics_l = df_topics.apply(lambda x: df_topics.columns[x == 1].tolist(), axis=1)
df_lang_l = df_languages.apply(lambda x: df_languages.columns[x == 1].tolist(), axis=1)

# # lista de lista para usar no apriori
df_mergex = []

for index in range(df.shape[0]):
    df_mergex.append(df_topics_l[index] + df_lang_l[index])
```

```python
from apyori import apriori
```

```python
rules = apriori(df_lang_l, min_support=0.009, min_confidence=0.2, min_lift=3, min_length=2)
# Support: number of transactions containing set of times / total number of transactions
# .      --> products that are bought at least 3 times a day --> 21 / 7501 = 0.0027
# Confidence: Should not be too high, as then this wil lead to obvious rules

results = list(rules)
results_df = pd.DataFrame(results)
results_df.head(60)
```

```python
rules = apriori(df_topics_l, min_support=0.004, min_confidence=0.2, min_lift=3, min_length=2)
results = list(rules)
results_df = pd.DataFrame(results)
results_df.head(60)
```

```python
rules = apriori(df_mergex, min_support=0.009, min_confidence=0.2, min_lift=3, min_length=2)
results = list(rules)
results_df = pd.DataFrame(results)
results_df.head(60)
```
