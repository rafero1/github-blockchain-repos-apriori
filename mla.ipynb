## Algoritmo

Primeiro, realizamos uma operação *drop* nos tópicos ou linguagens que só são usados uma vez em todo o dataset. Não será possível criar grupos com eles, então para regras de associação são irrelevantes

```python
# retorna todas as colunas com somatório 1
df_topics_cols_drop = df_topics.loc[:, df_topics.sum() == 1].columns
df_languages_cols_drop = df_languages.loc[:, df_languages.sum() == 1].columns

# dropa essas colunas
df_topics = df_topics.drop(columns=df_topics_cols_drop)
df_languages = df_languages.drop(columns=df_languages_cols_drop)
```

```python
from apyori import apriori
from mlxtend.frequent_patterns import apriori as mlx_apriori
from mlxtend.frequent_patterns import association_rules, fpgrowth, fpmax
```

Support:    Support refers to the default popularity of an item and can be calculated by finding number of 
            transactions containing a particular item divided by total number of transactions.
            Support is an important measure because a rule that has very low support may occur simply by 
            chance.

            Support(B) = (Transactions containing (B))/(Total Transactions)

Confidence: Confidence refers to the likelihood that an item B is also bought if item A is bought. 
            It can be calculated by finding the number of transactions where A and B are bought together, 
            divided by total number of transactions where A is bought. 

            Confidence(A -> B) = (Transactions containing both (A and B))/(Transactions containing A)

Lift:       Lift refers to the increase in the ratio of sale of B when A is sold. Lift(A –> B) can be 
            calculated by dividing Confidence(A -> B) divided by Support(B). Lift basically tells us that 
            the likelihood of buying A and B together is (x) times more than the likelihood of 
            just buying B. A Lift of 1 means there is no association between products A and B. 
            Lift of greater than 1 means products A and B are more likely to be bought together. 
            Finally, Lift of less than 1 refers to the case where two products are unlikely to be bought 
            together.

            Lift(A -> B) = (Confidence (A -> B))/(Support (B))

### Tópicos

```python
frequent_itemsets_t = mlx_apriori(df_topics, min_support=0.005, use_colnames=True)
rules_t = association_rules(frequent_itemsets_t, metric="confidence", min_threshold=0.8)
```

* Baixo suporte pois não temos muitos tópicos além de blockchain

```python
rules_t.loc[rules_t['lift'] >= 3].sort_values(['conviction'], ascending=False).head(60)
```

### Linguagens

```python
frequent_itemsets_l = mlx_apriori(df_languages, min_support=0.1, use_colnames=True)
rules_l = association_rules(frequent_itemsets_l, metric="confidence", min_threshold=0.5)
```

```python
rules_l.loc[rules_l['lift'] > 1].sort_values(['conviction'], ascending=False).head(60)
```
